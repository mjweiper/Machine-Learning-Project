tokyo_2020_df = pd.read_csv("2020 Tokyo Medals.csv")
tokyo_2020_df = tokyo_2020_df.sort_values(by=['Country'])
tokyo_2020_df = tokyo_2020_df.replace('United States of America', 'United States')
tokyo_2020_df = tokyo_2020_df.replace("People's Republic of China",'China')
tokyo_2020_df = tokyo_2020_df.replace('ROC','Russia')
tokyo_2020_df = tokyo_2020_df.replace('Republic of Korea','North Korea')
tokyo_2020_df = tokyo_2020_df.replace('Islamic Republic of Iran','Iran')
tokyo_2020_df = tokyo_2020_df.replace('ROC','Russia')

tokyo_2020_df

```python
xlsx = pd.ExcelFile('RawData/2020 Data for Medals.xlsx')
athletes_2020_df = pd.read_excel(xlsx, 'Athletes')
Pop_GDP_2020_df = pd.read_excel(xlsx, 'Total Data')
Pop_GDP_2020_df
```

```python
athletes_2020_df
```

```python
xlsx = pd.ExcelFile('RawData/Olympic coordinates.xlsx')
coordinates_df = pd.read_excel(xlsx)
coordinates_df
```

```python
tokyo_2020_df = Pop_GDP_2020_df.merge(athletes_2020_df, left_on='Country', right_on='Country')
tokyo_2020_df = tokyo_2020_df[["YearCountry","Country", "Year","Gold","Silver","Bronze","Total","Athletes","Population","GDP"]]
tokyo_2020_df
```

```python
tokyo_2020_df = tokyo_2020_df.merge(coordinates_df, how= 'outer', left_on='Country', right_on='Country')
tokyo_2020_df
```

```python
rio_2016_df = pd.read_csv("RawData/2016medalcount.csv")
rio_2016_df.head()
```

```python
rio_2016_df = rio_2016_df.merge(coordinates_df, how= 'outer',left_on='Country', right_on='Country')
rio_2016_df
```

```python
xlsx = pd.ExcelFile('RawData/2000_Olympics.xlsx')
sydney_2000_df = pd.read_excel(xlsx)
sydney_2000_df
```

```python
sydney_2000_df = sydney_2000_df.merge(coordinates_df, how= 'outer',left_on='Country', right_on='Country')
sydney_2000_df
```

```python
xlsx = pd.ExcelFile('RawData/2004_Olympics.xlsx')
athens_2004_df = pd.read_excel(xlsx)
athens_2004_df.head()
```

```python
athens_2004_df = athens_2004_df.merge(coordinates_df, how= 'outer',left_on='Country', right_on='Country')
athens_2004_df
```

```python
xlsx = pd.ExcelFile('RawData/2008_Olympics.xlsx')
beijing_2008_df = pd.read_excel(xlsx)
beijing_2008_df.head()
```

```python
beijing_2008_df = beijing_2008_df.merge(coordinates_df, how= 'outer',left_on='Country', right_on='Country')
beijing_2008_df
```

```python
xlsx = pd.ExcelFile('RawData/2012 Olympic coordinates.xlsx')
london_2012_df = pd.read_excel(xlsx)
london_2012_df.head()
```

```python
final_df = tokyo_2020_df.append(rio_2016_df)
final_df = final_df.append(london_2012_df)
final_df = final_df.append(beijing_2008_df)
final_df = final_df.append(athens_2004_df)
final_df = final_df.append(sydney_2000_df)
final_df
```

```python
final_df.to_csv ('final_df.csv', index = False, header=True)
final_df.to_excel ('final_df.xlsx', index = False, header=True)
```
